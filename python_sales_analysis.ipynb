**Q1. What was the best month for sales? How much was earned that month?**

```python
all_data['Quantity Ordered'][1070]
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'], format = '%m/%d/%y %H:%M', errors = 'coerce')
```

```python
quantity_empty = pd.to_numeric(all_data['Quantity Ordered'], errors = 'coerce').isna()
quantity_empty.sum()
```

```python
price_empty = pd.to_numeric(all_data['Price Each'], errors = 'coerce').isna()
price_empty.sum()
```

```python
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['sale'] = all_data['Price Each'] * all_data['Quantity Ordered']
all_data.head()
```

```python
all_data.info()
```

```python
all_data['Month'] = all_data['Order Date'].dt.month_name()
```

```python
all_data.head()
```

```python
all_data.loc[all_data['Month'] == 'January','Price Each']
```

```python
sales = {}
for i in month:
    prices = all_data.loc[all_data['Month'] == month[i],'Price Each'].astype('float64')
    quantity = all_data.loc[all_data['Month'] == month[i],'Quantity Ordered'].astype('float64')
    total_price = prices * quantity
    total_sales = total_price.sum()
    sales[i] = total_sales
```

```python
sales
```

```python
max_key = max(sales,key =sales.get)
```

```python
max_sale_month = month[max_key]
max_sale = sales[max_key]
print("The best month for sale is '%s' and sale of that month is %.2f"%(max_sale_month,max_sale))
```

```python
#ALTERNATE WAY OF DOING IT.
results = all_data.groupby('Month', sort = False).sum('sale')
results
```

```python
#plotting the result
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))

plt.bar(list(month.values()), results['sale'])
plt.xticks(list(month.values()),rotation = 45)
plt.xlabel('Months', fontsize = 14)
plt.ylabel('Sales in (in million USD)', fontsize = 14)
plt.title('Sales Analysis', fontsize = 16, fontweight = 'bold')
plt.show()
```

```python
#way 1
plt.figure(figsize=(12, 6))
plt.hist(grouped.index, weights=grouped['sale'], bins=24, edgecolor='black', alpha=0.8)

plt.xticks(range(24))
plt.xlabel("Hour of Day")
plt.ylabel("Sales")
plt.title("Histogram of Sales by Hour")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
# recommendation for ads is 10:00-12:00 and 19:00-21:00
```

```python
all_data.head()
```

```python
def hour_bucket(t):
    start_hour = str(t.hour).zfill(2) + ":00"
    end_hour = str((t.hour + 1) % 24).zfill(2) + ":00"
    return start_hour + " - " + end_hour

all_data['hour_interval'] = all_data['time'].apply(hour_bucket)

grouped_hour_interval = all_data.groupby('hour_interval').sum(numeric_only= True)

print(grouped_hour_interval)
```

```python
day_and_time_sale = all_data.groupby(['day','hour_interval']).sum(numeric_only= True)
```

```python
day_and_time_sale
```

```python
day_and_time_sale= day_and_time_sale.reset_index()
```

```python
#way2
#day_and_time_sale= day_and_time_sale.reset_index() - added as comment because already ran it once
days = day_and_time_sale['day'].unique()
plt.figure(figsize=(14, 8))

for day in days:
    subset = day_and_time_sale[day_and_time_sale['day'] == day]
    plt.bar(subset['hour_interval'], subset['sale'], alpha=0.75, label=day)

plt.xticks(rotation=45, ha='right')
plt.xlabel('Hour Interval')
plt.ylabel('Sales')
plt.title('Sales Distribution by Hour Interval for Each Day')
plt.legend()
plt.show()
# recommendation for ads is 10:00-12:00 and 19:00-21:00 and there is no much difference of day 
```

**Q4. What products are most often sold together?**

```python
all_data.head(10)
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df.head()
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df.head()
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
```

```python
df.head()
```

```python
#check solution ipynp file for fully unerstanding
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2))) #we can change the number from 2 to 3 and get 3  pairs that are baught together
print(count)
```

```python
count.most_common(10)
```

**Q5. What product sold the most? Why do you think it sold the most?**

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum(numeric_only=True)['Quantity Ordered']
```

```python
print(quantity_ordered)
```

```python
#plotting the result
plt.figure(figsize = (10,8))
products = [product for product,df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation = 'vertical')
plt.xlabel('Products', fontsize = 14)
plt.ylabel('Quantity Ordered', fontsize = 14)
plt.title('Product Demand', fontsize = 16, fontweight = 'bold')
plt.show()
```

```python
prices = all_data.groupby('Product').sum(numeric_only=True)['Price Each']
print(prices)
```

```python
#adding the prices onto the previous plot
fig,ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered,color = 'b')
ax2.plot(products,prices, 'r')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'b')
ax2.set_ylabel('Prices ($)', color = 'r')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)

plt.show()
```
